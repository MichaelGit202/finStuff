```python
from ib_async import *

# 1. This tells ib_async to safely patch into Jupyter's existing background loop
util.startLoop()

# 2. Initialize the client
ib = IB()

# 3. Connect to the local live gateway (port 4001)
# Note: In Jupyter, we use standard .connect() instead of .connectAsync()
print("Connecting...")
ib.connect('127.0.0.1', 4001, clientId=1,timeout=5)

# 4. Use free CBOE real-time data type
ib.reqMarketDataType(1)

# 5. Define contract and qualify it
contract = Stock('AAPL', 'CBOE', 'USD')
ib.qualifyContracts(contract)

# 6. Request the live market data ticker object
ticker = ib.reqMktData(contract)
print("Streaming started.")
```

```python

# Run this cell whenever you want to see the latest snapshot
print(f"Bid: {ticker.bid} | Ask: {ticker.ask} | Last Price: {ticker.last}")
```
